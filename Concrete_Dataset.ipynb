```python
plt.figure(figsize=(14, 6))
# Using seaborn to plot all columns on one scale (if scales differ widely, plot separately)
sns.boxplot(data=data)
plt.title('Box Plots for Outlier Detection')
plt.xticks(rotation=45) # Rotate labels if they overlap
plt.show()
```

```python
sns.pairplot(data, diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Change 'Cement', 'Strength', and 'Water' to your actual column names
sns.scatterplot(
    data=data,
    x='Cement_cmpo',  
    y='Concrete compressive strength(MPa, megapascals) ', 
    size='Water  (component 4)(kg in a m^3 mixture)',   # The dots get bigger as 'Water' increases
    hue='Water  (component 4)(kg in a m^3 mixture)',    # The color changes as 'Water' increases
    sizes=(20, 200), # Range of dot sizes
    palette='viridis',
    alpha=0.7       # Transparency to see overlapping dots
)
plt.title('Impact of Cement and Water on Strength')
plt.show()
```
