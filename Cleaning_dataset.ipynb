```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
df.head()
```

```python
df['Product_Category'].value_counts().plot(kind='bar')
```

```python
df['Product_Category'].value_counts().plot(kind='bar')
plt.ylim(2800,3100)
```

```python
df['Delivery_Status'].value_counts().plot(kind='bar')
```

```python
df['Review_Rating'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(12, 5))
# Group by Date to get daily total sales
daily_sales = df.groupby('Date')['Total_Sales_INR'].sum().reset_index()
sns.lineplot(data=daily_sales, x='Date', y='Total_Sales_INR', color='Red', linewidth=1)
plt.title('Daily Sales Trend', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Total Sales (INR)')
plt.show()
```

```python
plt.pie(df['Payment_Method'])
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Product_Category', y='Review_Rating', palette='coolwarm')
plt.title('Review Rating Spread by Category', fontsize=16)
plt.show()
```

```python
plt.plot(df['Review_Rating'], df['State'], label='Review_Rating') 
plt.plot(df['Review_Rating'], df['Quantity'], label='Quantity')
plt.legend() 
plt.show() 
```

```python
sns.histplot(data=df, x="Product_Category", hue="Review_Rating", multiple="stack")
```
