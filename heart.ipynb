### Visualization

```python
import seaborn.objects as so
```

```python
p1 = so.Plot(df, "RI", "Na")
p1.add(so.Dots())
```

```python
p = so.Plot(df, "Mg","Al").add(so.Dots())
p.facet("Type")
```

```python
(
    so.Plot(df, "Si", "Na", color="Type")
    .add(so.Dot(), so.Dodge(), so.Jitter(.2))
)
```

```python
(
    so.Plot(df, x="K", y="Ca")
    .add(so.Dot(pointsize=3), so.Shift(y=.2), so.Jitter(.2))
    .add(so.Dot(), so.Agg())
    .add(so.Range(), so.Est(errorbar=("se", 2)))
)
```

```python
(
    so.Plot(df, "RI", "Type")
    .add(so.Dots(), so.Jitter(.25))
)
```

```python
sns.pairplot(data=df, hue="Type")
```
